```python
import matplotlib.pyplot as plt
import numpy as np

# Read the .dat file with NumPy's loadtxt function
file_path = r"He_latpar_sweep.dat"

# Use numpy.loadtxt
data = np.loadtxt(
    file_path, 
    skiprows=1, 
    # delimiter defaults to any whitespace, which is ideal for .dat files
)

# Extract columns from the NumPy array
# data is a 2D array, so we select all rows (:) and the first column (0)
x = data[:, 1] 
# Select all rows (:) and the second column (1)
y = data[:, 0]  

# Create the plot
plt.plot(x, y, marker='o', linestyle='-')
plt.xlabel('Lattice Parameter a (A)')
plt.ylabel('Ground State Energy (eV)')
plt.title(' Spin Pol Ground States')
plt.grid(True)
plt.show()
```
